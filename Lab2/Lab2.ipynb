```python
import pandas as pd

original = pd.read_csv('https://raw.githubusercontent.com/johanavalos/MachineLearning/refs/heads/main/weather.csv')

original.head()
```

```python
original.columns.to_list()
```

```python
allowed_prefixes = ("DATE", "MONTH", "MAASTRICHT", "DE_BILT", "DUSSELDORF")
original = original[[col for col in original.columns if col.startswith(allowed_prefixes)]]
```

```python
original.columns.to_list()
```

```python
original.head()
```

```python
original.isna().sum()
```

```python
# List of region prefixes
regions = ['DE_BILT', 'DUSSELDORF', 'MAASTRICHT']

# Melt the DataFrame to unpivot the region-specific columns
df_melted = original.melt(id_vars=['DATE', 'MONTH'], var_name='feature', value_name='value')

# Extract REGION and weather feature (remove the region from the feature name)
df_melted['REGION'] = df_melted['feature'].str.extract(f"^({'|'.join(regions)})")
df_melted['feature'] = df_melted['feature'].str.replace(f"^({'|'.join(regions)})_", "", regex=True)

# Pivot to create a more structured table with REGION as a column
df_transformed = df_melted.pivot_table(index=['DATE', 'MONTH', 'REGION'], columns='feature', values='value').reset_index()
```

```python
df_transformed.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()
df_transformed['region_encoded'] = label_encoder.fit_transform(df_transformed['REGION'])

current_columns = df_transformed.columns.tolist()

cols = current_columns.copy()
cols.remove('region_encoded')

region_index = cols.index('REGION')

cols.insert(region_index + 1, 'region_encoded')

df_transformed = df_transformed[cols]

df_transformed.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Exclude non-numeric columns
numeric_df = df_transformed.select_dtypes(include=[np.number])

# Calculate the correlation matrix for numeric data only
corr_matrix = numeric_df.corr()

plt.figure(figsize=(16, 8))
# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Display the heatmap

plt.show()
```
